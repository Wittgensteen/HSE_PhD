## 1.3. Two class and two demension with correlation balanced classes


```python
def get_corr_1():
    dis1 = np.random.multivariate_normal(mean = [-2, 0, 2, 4], cov=[[1.0, 0.8, 0.5, 0.3],
                                                                    [0.8, 1.0, 0.0, 0.0],
                                                                    [0.5, 0.0, 1.0, 0.2],
                                                                    [0.3, 0.0, 0.2, 1.0]], size = 1000)  # Y==1
    dis2 = np.random.multivariate_normal(mean = [2, 1, 3, 5], cov=[[1.0, 0.1, 0.8, 0.2],
                                                                   [0.1, 1.0, 0.0, -0.2],
                                                                   [0.8, 0.0, 1.0, 0.0],
                                                                   [0.2, -0.2, 0.0, 1.0]], size = 1000)  # Y==2
    corr_1_1 = pd.DataFrame(dis1).rename(columns={0:'X1', 1:'X2', 2:'X3', 3:'X4' })
    corr_1_2 = pd.DataFrame(dis2).rename(columns={0:'X1', 1:'X2', 2:'X3', 3:'X4' })

    corr_1_1['Y']=1
    corr_1_2['Y']=2

    corr_set_1 = corr_1_1.append(corr_1_2)

    return  corr_set_1

corr_set_1 = get_corr_1()

corr_set_1.to_pickle('corr_set_1_dt')

```

# 3. DecisionTreeClassifier

```python
from sklearn.tree import DecisionTreeClassifier 
from sklearn.model_selection import train_test_split 
from sklearn import metrics 
from sklearn.metrics import roc_curve, auc

from sklearn.tree import export_graphviz
from sklearn.externals.six import StringIO  
from IPython.display import Image  
import pydotplus
import os     
```

```python
def check_DecisionTreeClassifier(data_frame):

    clf = DecisionTreeClassifier(criterion="entropy", max_depth=3)
    
    X = data_frame[['X1', 'X2', 'X3', 'X4']]
    y = data_frame['Y']
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
    model = clf.fit(X = X_train, y = y_train)
    print('Train Score: \n',model.score(X_train,y_train))
    y_pred = model.predict(X_test)
    comp = pd.DataFrame(y_test)
    comp['Y_pred'] = y_pred
    comp['Acc'] = np.where((comp['Y'] == comp['Y_pred']), 'True', 'False')
    print('Test Accuracy\n', comp['Acc'].value_counts())
    
#     dot_data = StringIO()
#     export_graphviz(clf, out_file=dot_data,  
#                 filled=True, rounded=True,
#                 special_characters=True,feature_names = X.columns,class_names=['1','2'])
#     graph = pydotplus.graph_from_dot_data(dot_data.getvalue())  
#     graph.write_png('tst.png')
#     Image(graph.create_png())
    
```
