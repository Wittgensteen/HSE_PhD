## 1.1. Two class and two demension without correlation balanced classes

```python
def get_normal_1():
    
    normal_1_x1y1 = np.random.normal(-2, 1, 1000)   # mu, sigma, n_rows
    normal_1_x1y2 = np.random.normal(2, 1, 1000)
    normal_1_x2y1 = np.random.normal(0, 1, 1000)
    normal_1_x2y2 = np.random.normal(0, 1, 1000)

    normal_2_x1y1 = np.random.normal(-1, 1, 1000)
    normal_2_x1y2 = np.random.normal(1, 1, 1000)
    normal_2_x2y1 = np.random.normal(1, 0.5, 1000)
    normal_2_x2y2 = np.random.normal(-1, 0.5, 1000)

    normal_3_x1y1 = np.random.normal(-2, 1, 1000)
    normal_3_x1y2 = np.random.normal(2, 1, 1000)
    normal_3_x2y1 = np.random.normal(0, 0.5, 1000)
    normal_3_x2y2 = np.random.normal(0, 0.5, 1000)

    normal_4_x1y1 = np.random.normal(-2, 1, 1000)
    normal_4_x1y2 = np.random.normal(2, 0.5, 1000)
    normal_4_x2y1 = np.random.normal(0, 0.5, 1000)
    normal_4_x2y2 = np.random.normal(0, 1, 1000)



    normal_1_df_1 = pd.DataFrame(columns= ['X1', 'X2', 'Y'])
    normal_1_df_2 = pd.DataFrame(columns= ['X1', 'X2', 'Y'])

    normal_2_df_1 = pd.DataFrame(columns= ['X1', 'X2', 'Y'])
    normal_2_df_2 = pd.DataFrame(columns= ['X1', 'X2', 'Y'])

    normal_3_df_1 = pd.DataFrame(columns= ['X1', 'X2', 'Y'])
    normal_3_df_2 = pd.DataFrame(columns= ['X1', 'X2', 'Y'])

    normal_4_df_1 = pd.DataFrame(columns= ['X1', 'X2', 'Y'])
    normal_4_df_2 = pd.DataFrame(columns= ['X1', 'X2', 'Y'])


    normal_1_df_1['X1'] = normal_1_x1y1
    normal_1_df_1['Y'] = 1
    normal_1_df_1['X2'] = normal_1_x2y1

    normal_1_df_2['X1'] = normal_1_x1y2
    normal_1_df_2['Y'] = 2
    normal_1_df_2['X2'] = normal_1_x2y2

    normal_2_df_1['X1'] = normal_2_x1y1
    normal_2_df_1['Y'] = 1
    normal_2_df_1['X2'] = normal_2_x2y1

    normal_2_df_2['X1'] = normal_2_x1y2
    normal_2_df_2['Y'] = 2
    normal_2_df_2['X2'] = normal_2_x2y2

    normal_3_df_1['X1'] = normal_3_x1y1
    normal_3_df_1['Y'] = 1
    normal_3_df_1['X2'] = normal_3_x2y1

    normal_3_df_2['X1'] = normal_3_x1y2
    normal_3_df_2['Y'] = 2
    normal_3_df_2['X2'] = normal_3_x2y2

    normal_4_df_1['X1'] = normal_4_x1y1
    normal_4_df_1['Y'] = 1
    normal_4_df_1['X2'] = normal_4_x2y1

    normal_4_df_2['X1'] = normal_4_x1y2
    normal_4_df_2['Y'] = 2
    normal_4_df_2['X2'] = normal_4_x2y2



    set1_1_1 = normal_1_df_1.append(normal_1_df_2).reset_index()
    set1_1_2 = normal_2_df_1.append(normal_2_df_2).reset_index()
    set1_1_3 = normal_3_df_1.append(normal_3_df_2).reset_index()
    set1_1_4 = normal_4_df_1.append(normal_4_df_2).reset_index()


    del set1_1_1['index']
    del set1_1_2['index']
    del set1_1_3['index']
    del set1_1_4['index']
    return set1_1_1, set1_1_2 ,set1_1_3, set1_1_4

set1_1_1, set1_1_2 ,set1_1_3, set1_1_4 = get_normal_1()

set1_1_1.to_pickle('set1_1_1')
set1_1_2.to_pickle('set1_1_2')
set1_1_3.to_pickle('set1_1_3')
set1_1_4.to_pickle('set1_1_4')
```

## 1.3. Two class and two demension with correlation balanced classes


```python
def get_corr_1():
    dis1 = np.random.multivariate_normal(mean = [-2,0], cov=[[1, 0.8],
                                                             [0.8, 1]], size = 1000)  # Y==1
    dis2 = np.random.multivariate_normal(mean = [2,0], cov=[[1, -0.4],
                                                             [-0.4, 1]], size = 1000)  # Y==2
    corr_1_1 = pd.DataFrame(dis1).rename(columns={0:'X1', 1:'X2' })
    corr_1_2 = pd.DataFrame(dis2).rename(columns={0:'X1', 1:'X2' })

    corr_1_1['Y']=1
    corr_1_2['Y']=2

    corr_set_1 = corr_1_1.append(corr_1_2)

    #____________________________________________________________________________________________#

    dis3 = np.random.multivariate_normal(mean = [1,2], cov=[[1, 0],
                                                             [0, 1]], size = 1000)  # Y==1
    dis4 = np.random.multivariate_normal(mean = [-2,1], cov=[[1, 0.7],
                                                             [0.7, 1]], size = 1000)  # Y==2
    corr_2_1 = pd.DataFrame(dis3).rename(columns={0:'X1', 1:'X2' })
    corr_2_2 = pd.DataFrame(dis4).rename(columns={0:'X1', 1:'X2' })

    corr_2_1['Y']=1
    corr_2_2['Y']=2

    corr_set_2 = corr_2_1.append(corr_2_2)
    return  corr_set_1, corr_set_2

corr_set_1, corr_set_2 = get_corr_1()

corr_set_1.to_pickle('corr_set_1')
corr_set_2.to_pickle('corr_set_2')

```

# 4. LinearDiscriminantAnalysis

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.model_selection import train_test_split

lda = LinearDiscriminantAnalysis()
```

```python
def check_lda(data_frame):
    """
    Inputs - dataframe
    Returns - Train Score, Train means, Test Accuracy table
    """
    lda = LinearDiscriminantAnalysis()
    
    X = data_frame[['X1', 'X2']]
    y = data_frame['Y']
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
    model = lda.fit(X = X_train, y = y_train)
    print('Train Score: \n',model.score(X_train,y_train))
    print('Train means_: \n',model.means_)
    y_pred = model.predict(X_test)
    comp = pd.DataFrame(y_test)
    comp['Y_pred'] = y_pred
    comp['Acc'] = np.where((comp['Y'] == comp['Y_pred']), 'True', 'False')
    print('Test Accuracy\n', comp['Acc'].value_counts())
    
```
